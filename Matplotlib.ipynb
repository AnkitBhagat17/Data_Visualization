```python
import matplotlib.pyplot as plt
```

#### Bar Plot ----> plt.bar()

```python
c=["r", "g", "b", "y", "m"] #Colors for the bars
x = ["Python", "Java", "C++", "JavaScript", "Ruby"]
y = [85,70, 60, 82, 55]
plt.bar(x,y,color='red',width=0.4, edgecolor = 'black',linewidth=1.5, label='Popularity') # linewidth for bar edges and edegcolor for bar edges
plt.xlabel("Programming Languages",fontsize = 10)
plt.ylabel("Popularity",fontsize = 10)
plt.title("Programming Languages and their Popularity",fontsize = 10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["Python", "Java", "C++", "JavaScript", "Ruby"]
y = [85, 70, 60, 82, 55]
z = [20, 30, 40, 50, 60]
width = 0.4
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.bar(p, y, color='red', width=width, edgecolor='black', linewidth=1.5, label='Popularity 1')
plt.bar(p1, z, color='yellow', width=width, edgecolor='black', linewidth=1.5, label='Popularity 2')
plt.xlabel("Programming Languages", fontsize=10)
plt.ylabel("Popularity", fontsize=10)
plt.title("Programming Languages and their Popularity", fontsize=10)
plt.xticks([j + width / 2 for j in p], x, rotation = 20)
plt.legend()
plt.show()
```
