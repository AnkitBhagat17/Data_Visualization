#### Pandas For Data Analysis
1) Pandas is a python package providing fast,flexible, and expressive data structures designed to make working with "relational" or "labeled" data both easy and intuitive.
 2) It was created by Wes McKinney in 2008

 Applications : 
 1) Easy Handling
 2) Size Mutability : Columns can be inserted and deleted from DataFrame and higher dimensional obejcts
 3) Automatic align the data for you in computations
 4) Powerful,flexible group by functionality.
 5) Flexible reshaping and pivoting of data sets.

 Pandas Series is a one-dimesnional labeled array capable of holding data of any type(integer,string,float,python objects,etc)

 The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet.

 The objects supports both integer and label-based indexing and provides a host of methods for performing operations involving the index.

##### Exploring Data in Pandas

##### read_csv() or read_excel() : This function to read data stored as a csv or excel file into a pandas.

```python
import pandas as pd

df = pd.read_excel("ESD.xlsx")
df
```

##### head() : It is used to see the first N rows of a DataFrame. N = 1,2,3,........no of rows u want to see

```python
df.head(5)
```

##### tail() :  It is used to see the Last N rows of a DataFrame. Where N = 1 ,2 ,3, 4,........no of rows u want to see

```python
df.tail(10)
```

##### info() : The method info() provides technical information about a DataFrame.

```python
df.info()  # It displays the data types,columns,etc
```

##### describe() : The describe() method provides a quick overview of the numerical data in a DataFrame.

```python
df.describe()
```

##### isnull().sum() : Used to identify the number of missing (null/NaN) values in each column of a DataFrame.

```python
df.isnull().sum()
```

##### Working with Missing Data 

```python
import pandas as pd
import numpy as np 

df = pd.read_csv("company1.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df
```

```python
df.replace(np.nan,30000)
```

```python
df["salary"] = df["salary"].replace(np.nan,24400)
print(df)
print("Mean : ", df["salary"].mean())
```

```python
df["salary"].mean()
```

```python
df.fillna(method= "bfill")
```

```python
df.fillna(method="ffill")
```

##### Columns Transformation


```python
import pandas as pd

data = pd.read_excel("ESD.xlsx")
data.loc[(data["Bonus %"] == 0), "GetBonus"] = "no bonus"
data.loc[(data["Bonus %"] > 0, "GetBonus")] = "bonus"
data.head(10)
```
