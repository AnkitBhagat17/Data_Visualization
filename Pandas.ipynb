#### Pandas For Data Analysis
1) Pandas is a python package providing fast,flexible, and expressive data structures designed to make working with "relational" or "labeled" data both easy and intuitive.
 2) It was created by Wes McKinney in 2008

 Applications : 
 1) Easy Handling
 2) Size Mutability : Columns can be inserted and deleted from DataFrame and higher dimensional obejcts
 3) Automatic align the data for you in computations
 4) Powerful,flexible group by functionality.
 5) Flexible reshaping and pivoting of data sets.

 Pandas Series is a one-dimesnional labeled array capable of holding data of any type(integer,string,float,python objects,etc)

 The axis labels are collectively called index. Pandas Series is nothing but a column in an excel sheet.

 The objects supports both integer and label-based indexing and provides a host of methods for performing operations involving the index.
