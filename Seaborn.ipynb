##### Seaborn 
It is a library used to make statistical graph or visualization in python

It is builds with Matplotlib and Pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [9,5,6,7,8,4,5,3]
```

### Matplotlib

```python
# Using Matplotlib
plt.plot(var1,var2)
plt.show()
```

### Seaborn

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [9,5,6,7,8,4,5,3]
```

```python
sns.lineplot(x=var1,y=var2)
plt.show()
```

##### Line Plot

```python
import pandas as pd

df = pd.DataFrame({"var1":var1,"var2":var2})
sns.lineplot(x='var1',y='var2',data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df2 = pd.read_csv("penguins.csv")
df2
```

#### Bar Plot ------> sns.barplot()

```python
df2['island']
```

```python
sns.barplot(x=df2.island, y=df2.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm",data=df2)
plt.figure(figsize=(6,5))
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm",data=df2,hue="sex")
plt.show()
```

##### Order

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df2 , hue="sex",order=order1)
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df2 , color='yellow')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df2,hue='sex',palette='Accent')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df2,hue='sex',palette='prism')
plt.show()
```

##### Histogram --------> sns.histplot()

```python
sns.histplot(df2['flipper_length_mm'])
plt.show()
```

##### Bins

```python
sns.histplot(df2['flipper_length_mm'],bins=[170,180,200,210,220,230,240])
plt.show()
```

##### Color

```python
sns.histplot(df2['flipper_length_mm'],bins=[170,180,200,210,220,230,240],color='orange')
plt.show()
```

##### Kernel Density Estimation plot (KDE)

```python
sns.displot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,color='orange')
plt.show()
```
