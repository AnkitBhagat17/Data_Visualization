## Central Limit Theorem

It States that when plotting a sample distributions of means the means of the sample will be equal to the population mean and the sample distribution will approach normal distribution with variance equal to standard error

There are a few assumptions behind the central limit theorem:
1. The samples should be independent of each other.
2. The sample size should be sufficiently large (usually n > 30 is considered a good rule of thumb).
3. The population from which the samples are drawn should have a finite mean and variance.
4. The distribution of the population can be normal or non-normal, but as the sample size increases, the sampling distribution of the sample mean will approach a normal distribution.

## Hypothesis Testing

1. Hypothesis testing is used to determine if there is enough evidence to reject a null hypothesis.
2. The null hypothesis(Ho) is a statement that there is no effect or no difference, and it is what we test against.
3. The alternative hypothesis(Ha) is a statement that there is an effect or a difference, and it is what we want to prove.  
4. The p-value is the probability of observing the data given that the null hypothesis is true.
5. A low p-value (typically less than 0.05) indicates that the null hypothesis can be rejected, suggesting that the alternative hypothesis may be true.


6. Types of Test for Hypothesis Testing
    1. Z-Test 
    2. T-Test 
    3. Chi-Square Test

7.Steps of Hypothesis Testing

        1. State null(Ho) and alternative(Ha/H1) hypothesis
            1) Ho must always contain equality(=)
            2) H1 always contains difference(not =, >,<)
        2. Choose level of Significance(alpha), left-tailed,right-tailed,two-tailed
        3. Find critical values
        4. Find test statistic
        5. Draw your conclusion


---

#### 🔹 1. **Z-Test**

✅ **When to use:**

* You know the **population standard deviation (σ)**.
* Sample size is **large (n ≥ 30)** (Central Limit Theorem ensures normality).
* Data is **continuous** and approximately **normally distributed**.

📌 **Applications:**

* Comparing a sample mean with a population mean.
* Comparing two population means (large samples).
* Testing population proportion.

---

#### 🔹 2. **T-Test**

✅ **When to use:**

* **Population standard deviation (σ) is unknown**.
* Sample size is **small (n < 30)** (but still valid for larger n too).
* Data is **continuous**, approximately **normal**.

📌 **Types of T-Tests:**

* **One-sample t-test** → Compare sample mean to a known value.
* **Independent two-sample t-test** → Compare means of 2 independent groups.
* **Paired t-test** → Compare means of the same group before/after (dependent samples).

---

#### 🔹 3. **Chi-Square Test (χ² Test)**

✅ **When to use:**

* Data is **categorical** (nominal/ordinal, not continuous).
* You want to test **association or independence** between variables, or **goodness of fit**.

📌 **Applications:**

* **Chi-square test of independence** → Are two categorical variables related?
  (e.g., Gender vs. Preference)
* **Chi-square goodness of fit test** → Does observed data fit an expected distribution?
  (e.g., rolling a die → fair or biased?).

---

#### 📝 Quick Summary Table:

| Test           | Data Type   | Sample Size           | σ Known? | Use Case                                 |
| -------------- | ----------- | --------------------- | -------- | ---------------------------------------- |
| **Z-Test**     | Continuous  | n ≥ 30                | Yes      | Compare means/proportions (large sample) |
| **T-Test**     | Continuous  | n < 30 (or σ unknown) | No       | Compare means (small sample / σ unknown) |
| **Chi-Square** | Categorical | Any                   | N/A      | Test independence / goodness of fit      |

# 📊 Hypothesis Testing Formulas

---

## 🔹 1. Z-Test

**One-Sample Z-Test**

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

**Two-Sample Z-Test**

$$
Z = \frac{(\bar{X}_1 - \bar{X}_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}
$$

---

## 🔹 2. T-Test

**One-Sample T-Test**

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

**Two-Sample Independent T-Test**

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

**Two-Sample T-Test (Pooled Variance)**

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{s_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
$$

$$
s_p^2 = \frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}
$$

**Paired T-Test**

$$
t = \frac{\bar{d}}{s_d / \sqrt{n}}
$$

---

## 🔹 3. Chi-Square Test

**General Formula**

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

**Degrees of Freedom**

- Goodness of Fit:  
$$
df = k - 1
$$

- Test of Independence:  
$$
df = (r-1)(c-1)
$$

<b>Z-Test Example</b>
<b>Example 1 :</b> A teacher claims that the mean score of studentd in his class is greater than 82 with a standard deviation of 20. If a sample of 81 students was selected with a mean score of 90.

```python
# Given 
import math
popu_mean = 82
std_dev = 20
sample_mean = 90
n = 81


Zcal = (sample_mean-popu_mean)/(std_dev/math.sqrt(n))
print(Zcal,f"Claim is Correct")
```

```python
import scipy.stats as st 
import numpy as np
# st.norm.ppf(0.95)

sample_mean_x = 90
pop_mean_u = 82
std_dev = 20
no_of_sample_n = 81
alpha = 0.05

ZCal = (sample_mean_x - pop_mean_u)/(std_dev/np.sqrt(no_of_sample_n))
print("Printing the Z Calculated Value : ",ZCal)

Ztable = st.norm.ppf(1-alpha)
print("Printing the value from Z-Table : ",Ztable)

if Ztable < Zcal:
    print("Ha is right")
else:
    print("Ho is right")
```

<b>T-test Example 1:</b> A manufacturer claims that the average weight of a bag of potato chips is 150 grams. A sample of 25 bags is taken , and the average weight is found to be 148 grams, with a standard deviation of 5 grams. Test the manufacturer's claim using a one-tailed t-test with a significance level of 0.05
