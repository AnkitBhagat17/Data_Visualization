Here’s a complete list of **commonly used string functions in Python** with **short descriptions** and examples:

---

### ✅ **1. `len()`** – Get length of string

```python
s = "Hello"
print(len(s))  # Output: 5
```

---

### ✅ **2. `lower()` / `upper()`** – Convert to lowercase / uppercase

```python
s = "Hello"
print(s.lower())  # "hello"
print(s.upper())  # "HELLO"
```

---

### ✅ **3. `strip()` / `lstrip()` / `rstrip()`** – Remove whitespace

```python
s = "  Hello  "
print(s.strip())   # "Hello"
print(s.lstrip())  # "Hello  "
print(s.rstrip())  # "  Hello"
```

---

### ✅ **4. `replace(old, new)`** – Replace substring

```python
s = "hello world"
print(s.replace("world", "Python"))  # "hello Python"
```

---

### ✅ **5. `split(separator)`** – Split string into list

```python
s = "a,b,c"
print(s.split(","))  # ['a', 'b', 'c']
```

---

### ✅ **6. `join(iterable)`** – Join list into string

```python
words = ['Hi', 'There']
print(" ".join(words))  # "Hi There"
```

---

### ✅ **7. `find(sub)` / `rfind(sub)`** – First/last occurrence index

```python
s = "hello world"
print(s.find("o"))   # 4
print(s.rfind("o"))  # 7
```

---

### ✅ **8. `startswith()` / `endswith()`**

```python
s = "python"
print(s.startswith("py"))  # True
print(s.endswith("on"))    # True
```

---

### ✅ **9. `count(sub)`** – Count occurrences

```python
s = "banana"
print(s.count("a"))  # 3
```

---

### ✅ **10. `capitalize()` / `title()`** – Capitalize first letter or title case

```python
print("hello".capitalize())  # "Hello"
print("my name is john".title())  # "My Name Is John"
```

---

### ✅ **11. `isalnum()` / `isalpha()` / `isdigit()`**

```python
"abc123".isalnum()  # True
"abc".isalpha()     # True
"123".isdigit()     # True
```

---

### ✅ **12. `casefold()`** – Like `lower()`, but stronger (for comparisons)

```python
"Straße".casefold() == "strasse"  # True
```

---

### ✅ **13. `format()` / f-strings** – String formatting

```python
name = "Ankit"
print("Hello, {}".format(name))  # Hello, Ankit
print(f"Hello, {name}")          # Hello, Ankit
```

---

### ✅ **14. `zfill(width)`** – Pad with zeros

```python
"42".zfill(5)  # "00042"
```

---

### ✅ **15. `encode()` / `decode()`** – Encode to bytes

```python
s = "hello"
b = s.encode()  # b'hello'
print(b.decode())  # 'hello'
```

````python
a = (1,2,45,45,45,"Ankit",False,3.145)
print(type(a))
print(a)
print(a.count(45))
print(a.index(45))


for i in a:
    print(i)
````


## ✅ **Tuple Methods**

Since tuples are **immutable**, they support only a few methods compared to lists.

### 1. `count()`

Returns the number of times a specified value appears in the tuple.

```python
t = (1, 2, 2, 3, 2)
print(t.count(2))   # Output: 3
```

---

### 2. `index()`

Returns the **first index** of the specified value.

```python
t = (10, 20, 30, 40)
print(t.index(30))  # Output: 2
```

---

## 🧠 **Tuple Operations and Properties**

### 3. **Length of Tuple**

```python
t = (1, 2, 3)
print(len(t))  # Output: 3
```

---

### 4. **Concatenation**

```python
t1 = (1, 2)
t2 = (3, 4)
print(t1 + t2)  # Output: (1, 2, 3, 4)
```

---

### 5. **Repetition**

```python
t = (1, 2)
print(t * 3)  # Output: (1, 2, 1, 2, 1, 2)
```

---

### 6. **Membership Test**

```python
t = (10, 20, 30)
print(20 in t)   # Output: True
print(40 in t)   # Output: False
```

---

### 7. **Tuple unpacking**

```python
t = (1, 2, 3)
a, b, c = t
print(a, b, c)  # Output: 1 2 3
```

---

### 8. **Nested Tuples**

```python
t = (1, (2, 3), 4)
print(t[1])      # Output: (2, 3)
print(t[1][0])   # Output: 2
```

---

### 9. **Slicing Tuples**

```python
t = (10, 20, 30, 40, 50)
print(t[1:4])  # Output: (20, 30, 40)
```

---

### 10. **Using Tuples as Dictionary Keys**

Because tuples are immutable, they can be used as keys in dictionaries.

```python
d = {(1, 2): "a tuple key"}
print(d[(1, 2)])  # Output: a tuple key
```

<h4 style="text-align:center"><b>Chapter-5 : Sets & Dictionary </b></h4>

<b>Dictionary</b>
1) Dictionary is a collection of keys-value pairs.
2) It is unordered , mutable, and it is indexed.
3) Cannot contain duplicates keys.

````python
marks = {
    "Ankit": 99,
    "Shubham": 88,
    "Rohan" : 45
}
print(marks,type(marks))
print(marks['Ankit'])
print(marks['Shubham'])
````

<b>Dict Methods</b>

1) items() - Returns a list of (key,value)tuples.
2) keys() - Returns a list containing dictionary's keys.
3) update() - Updates the dictionary with supplied key-value pairs
4) get() - Returns the value of the specified keys

````python
mydict = {} # Empty Dictionary
marks = {
    "Ankit": 99,
    "Shubham": 88,
    "Rohan" : 45
}
print(marks.items())
print(marks.keys())
print(marks.values())

marks.update({"Ankit": 100,"Prachi":99})
print(marks)

print(marks.get("Ankit"))
````

---

### 🔹 1. `dict.clear()`

**Removes all items** from the dictionary.

```python
my_dict = {"a": 1, "b": 2}
my_dict.clear()
print(my_dict)  # Output: {}
```

---

### 🔹 2. `dict.copy()`

Returns a **shallow copy** of the dictionary.

```python
original = {"x": 10, "y": 20}
copy_dict = original.copy()
print(copy_dict)  # Output: {'x': 10, 'y': 20}
```

---

### 🔹 3. `dict.get(key, default=None)`

Returns the value for the specified key. If the key is not found, returns the default value.

```python
d = {"name": "Ankit"}
print(d.get("name"))        # Output: Ankit
print(d.get("age", 20))     # Output: 20
```

---

### 🔹 4. `dict.items()`

Returns a **view object** of the dictionary's `(key, value)` pairs.

```python
d = {"a": 1, "b": 2}
print(d.items())  # Output: dict_items([('a', 1), ('b', 2)])
```

---

### 🔹 5. `dict.keys()`

Returns a view of all the **keys**.

```python
d = {"a": 1, "b": 2}
print(d.keys())  # Output: dict_keys(['a', 'b'])
```

---

### 🔹 6. `dict.values()`

Returns a view of all the **values**.

```python
d = {"a": 1, "b": 2}
print(d.values())  # Output: dict_values([1, 2])
```

---

### 🔹 7. `dict.pop(key, default=None)`

Removes the key and returns its value. If key not found, returns `default`.

```python
d = {"x": 100, "y": 200}
print(d.pop("x"))        # Output: 100
print(d.pop("z", 0))     # Output: 0
```

---

### 🔹 8. `dict.popitem()`

Removes and returns the **last inserted (key, value)** pair.

```python
d = {"a": 1, "b": 2}
print(d.popitem())  # Output: ('b', 2)
```

---

### 🔹 9. `dict.update([other])`

Updates the dictionary with elements from another dict or iterable.

```python
d = {"a": 1}
d.update({"b": 2, "c": 3})
print(d)  # Output: {'a': 1, 'b': 2, 'c': 3}
```

---

### 🔹 10. `dict.setdefault(key, default=None)`

Returns the value of the key. If key does not exist, inserts it with the default value.

```python
d = {"name": "Ankit"}
d.setdefault("age", 20)
print(d)  # Output: {'name': 'Ankit', 'age': 20}
```

---

### 🔹 11. `dict.fromkeys(keys, value=None)`

Creates a new dictionary from a sequence of keys with the same value.

```python
keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)
print(new_dict)  # Output: {'a': 0, 'b': 0, 'c': 0}
```
