##### 🔸 Problem 1: Reverse a List

```python
# Input [1,2,3,4] 
# Output [4,3,2,1]
arr = [1,2,3,4]
def reverse_list(arr):
    return arr[::-1]
reverse_list(arr)
```

##### 🔸 Problem 2: Find the Maximum Element

```python
# Input [3,2,6,2]
#  Output = 6
arr=[3,2,6,2]
def max_element(arr):
    return max(arr)

max_element(arr)
```

##### 🔸 Problem 3: Remove Duplicates from List

```python
# Input [1,2,2,3,3,4]
# Output [1,2,3,4]
arr=[1,2,2,3,3,4]
def remove_duplicates(arr):
    return list(set(arr)) # set(arr) In Python, a set is an unordered collection of unique elements.
                          #So when you convert a list to a set, duplicates are automatically removed.
remove_duplicates(arr)
```

##### PART 2: STRINGS
✅ What is a String?
-> A string is a sequence of characters.

Strings in Python are immutable (you can't change a character directly).

| Operation       | Code Example             | Description        |
| --------------- | ------------------------ | ------------------ |
| Length          | `len(s)`                 | Length of string   |
| Indexing        | `s[0]`, `s[-1]`          | Access characters  |
| Slicing         | `s[1:4]`                 | Substrings         |
| Reversing       | `s[::-1]`                | Reverse string     |
| Upper / Lower   | `s.upper()`, `s.lower()` | Change case        |
| Check substring | `"abc" in s`             | Boolean check      |
| Replace         | `s.replace("a", "b")`    | Replace characters |


```python
# Example
s = 'datastructures'
print("The length of the string is : ",len(s))
print("Slicing the string : ",s[0:4])
print("Reversing the Sting : ",s[::-1])
print("Finding the word is awailable in given String : ","data" in s)
print("String In UpperCase : ", s.upper())
print("String in LowerCase: ",s.lower())
print("Replacing the letters:",s.replace("s","d"))
```

🔸 Problem 2: Check Palindrome
Input: "madam"
Output: True

```python
str = "madam"
def is_pallindrome(str):
    return str == str[::-1]
is_pallindrome(str)
```
