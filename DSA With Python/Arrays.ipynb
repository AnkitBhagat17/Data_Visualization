1) List(Arrays) : A list is a dynamic array in Python.
2) It can store elements of different data types.
3) Lists are mutable (we can change them after creation).


List Operartions Are : 
1) Append - Add element to end.
2) Insert - Insert at Index
3) Remove - Remove 1st occurence
4) Pop - Remove Last
5) Index - Find Index of Value
6) Sort - Sort In Ascending Order
7) Reverse - Reverse List
